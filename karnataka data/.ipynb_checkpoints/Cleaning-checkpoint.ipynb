```python
import numpy as np
import pandas as pd
import os
```

```python
files = [x for x in os.listdir('.') if 'sbm' in x]
```

```python
csvs = [pd.read_csv(x) for x in files]
```

```python
final = pd.concat(csvs, axis=0)
```

```python
final.reset_index(inplace = True)
```

```python
final.drop('index', axis=1, inplace=True)
```

```python
final
```

```python
final.to_csv('karnataka_sanitation', index=False)
```

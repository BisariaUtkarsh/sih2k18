```python
lulc = gpd.read_file('Kurudu_LULC.shp')
lulc[lulc['dscr1'] == 'Agricultural land']
```

```python
base = lulc[(lulc['dscr1'] == 'Agricultural land') & (lulc['dscr2'] == 'Crop land')].plot()
lulc[(lulc['dscr1'] == 'Agricultural land') & (lulc['dscr2'] == 'Agriculture plantation')].plot(ax=base, color='g')
lulc[(lulc['dscr1'] == 'Agricultural land') & (lulc['dscr2'] == 'Crop land')].plot()
```
